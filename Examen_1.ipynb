```python
def up(x,n):
  import numpy as np
  #Creamos un array para almacenar los valores de las primeras n funciones de Bessel evaluadas en x
  j=np.zeros((n+2),float)
  #Casluclamos los valores que cononcemos, lo valores de las funciones de Bessel 0 y 1
  j[0]=np.sin(x)/x
  j[1]=np.sin(x)/x**2-np.cos(x)/x
  #Si se pide el valor de una funciÃ³n de Bessel de un orden mas grande que 1 se calcula con la relaciÃ³n de recursiÃ³n
  if n>1:
    for k in range(1,n):
      j[k+1]=((2*k+1)/x)*j[k]-j[k-1]
  return j[n]

import numpy as np
import matplotlib.pyplot as plt

#Creamos dos arrays con los valores a graficar
x=np.arange(0.25,40,0.1)
y=np.zeros(len(x),float)

for i in range(1,len(np.arange(0.25,40,0.1))):
  y[i]=up(x[i],10)

plt.plot(x,y)
plt.show()
```
